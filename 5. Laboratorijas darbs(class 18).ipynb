Skaitliska integresana

```python
from numpy import random,sin 
from matplotlib import pyplot as plt
#mainigo diferencesana
N = 10000

a = 0
b = 5
c = 1

x= random.uniform(a, b, N)
y = random.uniform(a, c, N)
#kods kurs vizualize tas kas ir izdarits
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.title('Funkcija sin x^2 un tās integrālis')
plt.plot(x, y, 'ko')

#formula kas aprekina sin(x*x) integrali
N1 = 1

for i in range(N):
    if y[i] < (sin(x[i]*x[i] ))/ 2:
        plt.plot(x[i], y[i], 'go')
        N1 = N1 + 1
    else:
        plt.plot(x[i], y[i], 'ro')
        
plt.show()

#lauka prekinasana
S = (b - a) * (c - a)
jaunaS = float(S * N1 / N)
print("Laukuma skaitliskā vertība:", )
print(jaunaS)
```
