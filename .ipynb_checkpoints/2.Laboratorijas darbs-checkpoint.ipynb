```python
#<<1B1.py>>
#zimejam funkciju Cos
from numpy import *  #Importē skaitlisko metožu bibliotēkas funkcijas
x= linspace(0,7,70) # Trešais arguments ir ģenerējamo elementu skaits
y= cos(x)

import matplotlib.pyplot as plt
plt.grid() #Uz grafika izveidojam tīklu
plt.xlabel('x') #Izveidojam uzrakstu uz horiontālās ass
plt.ylabel('f(x)') #Izveidojam uzrakstu uz vertikālās ass
plt.title('Funkcija $cos(x)$') #Izveidojam grafiku kopumā raksturojošo uzrakstu
plt.plot(x, y, color = "#610B4B") #Uzzīmējam funkcionālās sakarības grafiku un izvelamies to krasu
plt.show() #Uzzīmēto, bet neredzamo grafiku parāda uz ekrāna formas


#zimejam funkciju Sin
from numpy import *
x= linspace(0,7,70)
y2= sin(x)

import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $sin(x)$')
plt.plot(x, y2, color = "#610B4B")
plt.show()
```

```python
 K#<<1B2.py>>
import sys
sys.path.append('/usr/local/anaconda3/lib/python3.6/site-packages')

from numpy import *
x = linspace(0, 7, 70)
y = sin(x)

import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $sin(x)$')
plt.plot(x, y, color='#FF0000')

y1=x
plt.plot(x, y1, color='#04B45F')

y2=x - x*x*x/(1*2*3)
plt.plot(x, y2, color='#610B4B')

y3=x - x*x*x/(1*2*3)+x*x*x*x*x/(1*2*3*4*5)
plt.plot(x, y3, color='#FF33CC')

y4=x - x*x*x/(1*2*3)+x*x*x*x*x/(1*2*3*4*5)-x*x*x*x*x*x*x/(1*2*3*4*5*6*7)
plt.plot(x, y4, color='#663300')
plt.show()
```

```python
# -*- coding: utf-8 -*-
x = [0, 1, 2, 3]
y = [0, 2, 1, 2]
N = len(x)
legend = []
from matplotlib import pyplot as plt
#print(plt.__doc__)
# grafika ass iestatījumi
plt.axis([-1, 4, -1, 3])
# režģa ieslēgšana
plt.grid()
# x ass apzīmējums
plt.xlabel("$x$")
# y ass apzīmējums
plt.ylabel("$y$")
# grafika nosaukums
plt.title("$f(x)$ funkcijas grafiks")
# zīmēšana ar noklusētiem iestatījumiem
plt.plot(x,y)
legend.append("zīmēšana ar noklusētiem iestatījumiem")
# zīmēšana ar marķieriem
plt.plot(x[0],y[0],'gh',markersize = 10)
legend.append("zīmēšana ar marķieriem, 1. piemērs")
plt.plot(x[1:N],y[1:N],'ro',markersize = 5)
legend.append("zīmēšana ar marķieriem, 2. piemērs")
# leģendas (raksturlīķņu nosaukumu) pievienošana
#print(plt.legend.__doc__)
plt.legend(legend, loc = 3, fancybox=True, framealpha=0.5, facecolor="green")
# grafika publicēšana
plt.show()
```

```python
# -*- coding: utf-8 -*-
from numpy import sin, linspace
def f1(x):
    return x
def f2(x):
    return f1(x) - x**3/(1*2*3)
def f3(x):
    return f2(x) + x**5/(1*2*3*4*5)
def f4(x):
    return f3(x) - x**7/(1*2*3*4*5*6*7)
x = linspace(0,4,41)
y = sin(x); y1 = f1(x); y2 = f2(x); y3 = f3(x); y4 = f4(x)
legend = []
from matplotlib import pyplot as plt
plt.grid()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.title("$sin(x)$ funkcija un tās izvirzījumi rindā")
plt.plot(x,y,color = "#000000"); legend.append("$sin(x)$")
plt.plot(x,y1,color = "#FF0000"); legend.append("$x$")
plt.plot(x,y2,color = "#00FF00"); legend.append("$x - x^3/3!$")
plt.plot(x,y3,color = "#0000FF"); legend.append("$x - x^3/3! + x^5/5!$")
plt.plot(x,y4,color = "#FFFF00"); legend.append("$x - x^3/3! + x^5/5! - x^7/7!$")
plt.legend(legend, loc = 3)
plt.show()
```
